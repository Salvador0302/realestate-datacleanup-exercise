#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
df_cinturon_sur['price_normalized'] = df_cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))

plt.figure(figsize=(10, 6))
for poblacion in df_cinturon_sur['level5'].unique():
    data = df_cinturon_sur[df_cinturon_sur['level5'] == poblacion]['price_normalized']
    plt.hist(data, bins=15, alpha=0.5, label=poblacion)

plt.title('Histogramas de Precios Normalizados por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(title='Poblaciones')
plt.grid(True)

plt.show()
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
df_poblaciones = df_cinturon_sur[df_cinturon_sur['level5'].isin(poblaciones)]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for i, poblacion in enumerate(poblaciones):
    ax = axes[i//2, i%2]
    df_poblacion = df_poblaciones[df_poblaciones['level5'] == poblacion]
    ax.scatter(df_poblacion['surface'], df_poblacion['price'], alpha=0.5)
    ax.set_title(f'Dispersión de precios vs superficie en {poblacion}')
    ax.set_xlabel('Superficie')
    ax.set_ylabel('Precio')

plt.tight_layout()

plt.show()
```
